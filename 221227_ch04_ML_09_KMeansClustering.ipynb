# 고객관련 데이터셋

```python
# 주요 라이브러리 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# https://raw.githubusercontent.com/bigdata-young/ai_26th/main/data/example_cluster.csv
df = pd.read_csv('https://raw.githubusercontent.com/bigdata-young/ai_26th/main/data/example_cluster.csv')
df
```

```python
# 산점도 - 3지점에 모여 있음 => 묶음을 클러스터라고 함 -> 3개 
sns.scatterplot(x='var_1', y='var_2', data=df) # var_1, var_2을 x,y로 그림을 그림
plt.show()
```

```python
# k평균 군집화(K Means Clustering) - 라이브러리 사용
from sklearn.cluster import KMeans
```

```python
# n_clusters는 나눠줄 클러스터(묶음)의 수
kmeans_model = KMeans(n_clusters=3, random_state=27)
```

```python
# 지도 -> X,y
# 비지도 -> X
# 지도 학습은 원래 X를 통해 y를 예측하는 형태였음
# 비지도 학습은 y값이 없음
kmeans_model.fit(df)
```

```python
# 예측
pred = kmeans_model.predict(df)
pred
```

```python
# label
df['label'] = pred # KMeans 산출한 그룹들ㅇ르 label로 투입
```

```python
# 산점도2 - label 적용한
sns.scatterplot(x='var_1', y='var_2', data=df, hue='label') # var_1, var_2을 x,y로 그림을 그림
plt.show()
```

```python
# 산점도2 - label 적용한
sns.scatterplot(x='var_1', y='var_2', data=df, hue='label', palette='rainbow') # var_1, var_2을 x,y로 그림을 그림
plt.show()
```

```python
# 이너셔(inertia)
# 각 그룹에서 중심과 각 그룹에 해당하는 데이터 간의 거리에 대한 합
kmeans_model.inertia_ # 각 그룹에 3090을 중심으로 뭉쳐있다
```

```python
# k가 엄청 많을 경우
kmeans_model2 = KMeans(n_clusters=500, random_state=27)
kmeans_model2.fit(df)
kmeans_model2.inertia_
```

```python
df['label2'] = kmeans_model2.predict(df)
```

```python
df.label2
```

```python
# 산점도2 - label 적용한
sns.scatterplot(x='var_1', y='var_2', data=df, hue='label2', palette='rainbow') # var_1, var_2을 x,y로 그림을 그림
plt.show()
```

```python
# 적절한 k값 설정 -> 엘보우 기법
df = df.drop(['label','label2'], axis=1)
df.columns
```

```python
# inertia 거리의 변화 폭이  둔화되는 지점-> 꺽이는 지점
distance = []
for k in range(2,10): # 2 ~9까지 반복
  k_model = KMeans(n_clusters=k)
  k_model.fit(df) # fit만 시켜도 inertia 생성
  distance.append(k_model.inertia_)
```

```python
distance
```

```python
# 선 차트
sns.lineplot(x=range(2,10), y=distance)
# 엘보우: 직관적이고 편리하기는 하나, 객관적인 건 아님 => 사람이 판단
```

# 이상거래 데이터

```python
# 이상거래 데이터 -> 단순화
# https://raw.githubusercontent.com/bigdata-young/ai_26th/main/data/customer.csv
df = pd.read_csv('https://raw.githubusercontent.com/bigdata-young/ai_26th/main/data/customer.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
# cc_num : credit card -> 카드번호 -> 특정고객
df.cc_num.nunique()
```

```python
# category 별로 고객들이 어떻게 소비하나 - 카테고리별로 분류됨
customer = pd.get_dummies(df, columns=['category'])
customer.info()
```

```python
# iloc => 정수 인덱스로 행, 열을 가져오겠다 (인덱서)
# 전체 [행,열] => 전체 행 중에 2번째 이후 (경계포함)열들만 표시 하겠다
customer.iloc[:,2:]
```

```python
for col_name in customer.iloc[:,2:].columns: # 범주들 이름으로 for문(반복)
    # 트랜잭션이 결제
    # print(col_name)
    # 기존에 있는 컬럼(열)에 있는 새로운 값 연산
    customer[col_name] = customer[col_name] * customer['amt']
    # 특정 범주(카테고리 포함 여부) : coustomer[col_name]
    # 소비액 : customer['amt']
    # -> 특정 카테고리에 대한 소비액
    # 한 행 -> 거래 => 열 -> 거래가 속한 범주의 열이라면 amt랑 곱해진 소비액 나올 거고,
    # 나머지 소속되지 않은 경우에 -> 0만 있을 것
```

```python
customer
```

```python
# cc_num 별로 합계를 해주겠다
# 카드번호별로 groupby -> sum()
customer_agg = customer.groupby('cc_num').sum()
customer_agg
# cc_num별 총 소비액인 amt
# 각기 category별 cc_num의 소비액인 각자 열
# -> 총 소비 또는 각 범주별 소비를 통해서 '분류', '군집화'가 가능
```

```python
customer.info()
```

```python
# 스케일러 -> 표준화 스케일러(Standard Scaler)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# 스케일링을 하고 나서 데이터
scaled = scaler.fit_transform(customer_agg)
scaled # numpyp array
```

```python
scaled_df = pd.DataFrame(scaled, columns=customer_agg.columns,
                         index=customer_agg.index)
```

```python
scaled_df
```

```python
# 엘보우 기법
distance = []
for k in range(2,10) : 
  k_model = KMeans(n_clusters=k, random_state=12)
  k_model.fit(scaled_df)
  distance.append(k_model.inertia_)
```

```python
sns.lineplot(x=range(2,10), y=distance)
```

```python
# 실루엣 계수
# - 이너셔(inertia) : 그룹(클러스터)의 중심으로 부터 다른 요소들간의 거리
# - 실루엣 계수 : 클러스터 내부의 평균거리 + 최근접한 다른 클러스터 데이터와의 평균 거리
from sklearn.metrics import silhouette_score
```

```python
silhoette = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k, random_state=12)
    # k_model.fit(scaled_df)
    # k_model.inertia_
    labels = k_model.fit_predict(scaled_df)
    # 실제의 데이터 값과, 분류된 라벨을 통해서 클러스터 내부 및 클러스터 간 거리를 계산
    silhoette.append(silhouette_score(scaled_df, labels)) # labels는 어떤 기준으로 분류되었는가?
```

```python
silhoette
```

```python
# 실루엣 계수는 더 클 수록 좋은 분류를 하고 있는 것
sns.lineplot(x=range(2,10), y=silhoette) # 4가 제일 높은 값
```

```python
k_model = KMeans(n_clusters=4, random_state=12)
labels = k_model.fit_predict(scaled_df)
```

```python
scaled_df['label'] = labels
```

```python
scaled_df
```

```python
scaled_df_means = scaled_df.groupby('label').mean() # 스케일별로 평균치
scaled_df_means
# 3번 그룹이 제일 많은?
```

```python
scaled_df_count = scaled_df.groupby('label').count().amt
scaled_df_count # 0이 많음
```

```python
scaled_df_count = scaled_df_count.rename('count')
```

```python
scaled_df_all = scaled_df_means.join(scaled_df_count)
scaled_df_all
```
